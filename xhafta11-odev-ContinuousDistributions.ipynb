```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# SÜREKLİ OLASILIK DAĞILIMLARI
```

```python
# NORMAL DAĞILIM (SÜREKLİ)

mu=0 
sigma=1 
x=np.arange(-5,5,0.1) 
y=stats.norm.pdf(x,0,1)

plt.plot(x,y) 
plt.title('Normal:  $\mu$=%.1f , $\sigma^2$=%.1f' % (mu,sigma))
plt.xlabel('x')   
plt.ylabel('Olasılık Yoğunluğu') 
plt.show()
```

```python
# BETA DAĞILIMI (SÜREKLİ)

a=0.5 
b=0.5 
x=np.arange(0.01 , 1, 0.01) 
y=stats.beta.pdf(x,a,b) 

plt.plot(x,y) 
plt.title('Beta: a=%.1f , b=%.1f' % (a,b))
plt.xlabel('x')   
plt.ylabel('Olasılık Yoğunluğu') 
plt.show()
```

```python
# ÜSTEL DAĞILIM (SÜREKLİ)

lambd=0.5 
x=np.arange(0, 15, 0.1) 
y=lambd * np.exp(-lambd * x)

plt.plot(x,y) 
plt.title('Exponential (Üstel) : $\lambda$ = %.2f' % lambd)
plt.xlabel('x')   
plt.ylabel('Olasılık Yoğunluğu') 
plt.show()
```

```python
# GAMMA DAĞILIMI (SÜREKLİ)

from scipy.stats import gamma

k_values = [1, 2, 3, 5]
theta_values = [2, 1, 1, 0.5]
linestyles = ['-', '--', ':', '-.']
x = np.linspace(1E-6, 10, 1000)

fig, ax = plt.subplots(figsize=(5, 3.75))

for k, t, ls in zip(k_values, theta_values, linestyles):
    dist = gamma(k, 0, t)
    plt.plot(x, dist.pdf(x), ls=ls, label=r'$k=%.1f,\ \theta=%.1f$' % (k, t))

plt.xlim(0, 10)
plt.ylim(0, 0.45)

plt.title('Gamma')
plt.xlabel('x')
plt.ylabel('Olasılık Yoğunluğu') 


plt.legend(loc=0)
plt.show()
```

```python
# WEİBULL DAĞILIMI (SÜREKLİ)

from scipy.stats import dweibull

k_values = [0.5, 1, 2, 2]
lam_values = [1, 1, 1, 2]
linestyles = ['-', '--', ':', '-.', '--']
mu = 0
x = np.linspace(-10, 10, 1000)

fig, ax = plt.subplots(figsize=(5, 3.75))

for (k, lam, ls) in zip(k_values, lam_values, linestyles):
    dist = dweibull(k, mu, lam)
    plt.plot(x, dist.pdf(x), ls=ls, label=r'$k=%.1f,\ \lambda=%i$' % (k, lam))

plt.xlim(0, 5)
plt.ylim(0, 0.6)

plt.title('Weibull')
plt.xlabel('x')
plt.ylabel('Olasılık Yoğunluğu') 

plt.show()
```
